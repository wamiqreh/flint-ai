# Task Quickstart
Submit a task and wait for completion using the Python SDK.

```python
import asyncio
from flint_ai import AsyncOrchestratorClient

async def run():
    client = AsyncOrchestratorClient('http://localhost:5156')
    try:
        task_id = await client.submit_task('dummy', 'hello from notebook')
        task = await client.wait_for_task(task_id)
        return task.model_dump()
    finally:
        await client.close()

await run()
```
